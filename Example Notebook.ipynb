```python
from utils import Data

models = ['ReynoldsStress']
grids = [50, 100, 200]
Re_numbers = [300, 1200]
data = Data(Re_numbers, models, grids)
```

```python
data.plot_residuals(models[0])
```

# MFR

## Errors

```python
# Percentage error
for model in models:
    for Re in Re_numbers:
        for grid in grids:
            mfr = data.mfr[model][Re][grid]
            dns_mfr = data.dns_mfr[Re]
            print(f'Model = {model}, Re_tau = {Re}, Grid = {grid} x {grid}, MFR ERROR: {round(100 * (mfr - dns_mfr) / dns_mfr, 2)}%')
            data.mfr_plot[model][Re][grid].plot(title=f'Re_tau = {Re}, {grid} x {grid}')
```
