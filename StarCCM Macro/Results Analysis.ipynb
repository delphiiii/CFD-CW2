# Utils for plots and data

```python
from utils import Data

# Set the Re_tau numbers for analysis
Re_numbers = [300]

models = ['Reynolds stress turbulence']
grids = [50, 100, 200]

data = Data(Re_numbers, models, grids)
```

# Mass Flow Rate

```python
data.dns_mfr
```

```python
data.mfr  # Broken down into model, Re_tau and gridsize
```

```python
# Percentage error
mfr_200 = data.mfr[models[0]][300][200]
dns_mfr = data.dns_mfr[300]
print(f'{round(100 * (mfr_200 - dns_mfr) / dns_mfr, 2)}%')
```
